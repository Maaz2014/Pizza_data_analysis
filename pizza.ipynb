#### Using the AND operator to satisfy two conditions at once.
#### If we want to look at places where city is los angeles and price range max >10.

```python
pizza_priceRangeMax=pizza['priceRangeMax']>10
pizza_city=pizza['city']=='Los Angeles'
pizza[pizza_priceRangeMax & pizza_city]
```

#### Using the OR operator to satisfy two conditions at once.
#### If we want to look at places where city is los angeles or price range max >15.

```python
pizza_priceRangeMax=pizza['priceRangeMax']>15
pizza_city=pizza['city']=='Los Angeles'
pizza[pizza_priceRangeMax | pizza_city] #checking the or condition
```

#### Now , if we want to drop longitude and latitude columns , it is going to be like this by using the drop command.

```python
pizza=pizza.drop(labels=['longitude','latitude'],axis=1)
```

#### Dealing with missing values

```python
#now,if I want to update the null values of a specific value with mode or mean or whatever we want as per the query
pizza['menus.amountMax']=pizza['menus.amountMax'].fillna((pizza['menus.amountMax'].mean()))
```

```python
pizza['priceRangeMin']=pizza['priceRangeMin'].fillna((pizza['priceRangeMin'].mean()))
```

```python
pizza['priceRangeMax']=pizza['priceRangeMax'].fillna((pizza['priceRangeMax'].mean()))
```

```python
pizza['menus.currency']=pizza['priceRangeCurrency'].fillna((pizza['priceRangeCurrency'].ffill()))
```

```python
pizza['menus.description']=pizza['menus.description'].fillna((pizza['menus.description'].ffill()))
```

```python
pizza['postalCode']=pizza['postalCode'].fillna((pizza['postalCode'].ffill()))
```

```python
pizza['priceRangeCurrency']=pizza['priceRangeCurrency'].fillna((pizza['priceRangeCurrency'].ffill()))
```
